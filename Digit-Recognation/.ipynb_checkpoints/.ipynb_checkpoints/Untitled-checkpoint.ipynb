```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from random import randrange
import os
import cv2
```

```python
mnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train.shape
```

```python
y_train
```

```python
numbers = ["zero",
          "um",
          "dois",
          "tres",
          "quatro",
          "cinco",
          "seis",
          "sete",
          "oito",
          "nove",]
```

```python
numbers[y_train[0]]
```

```python
n = randrange(60000)

plt.imshow(X_train[n], cmap='gray')
numbers[y_train[n]]
```

```python
X_train = X_train.astype('float32') / 255
X_test = X_test.astype('float32') / 255
```

```python
model = tf.keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[28,28]))
model.add(keras.layers.Dense(250, activation="relu"))
model.add(keras.layers.Dense(100, activation="relu"))
model.add(keras.layers.Dense(10, activation="softmax"))
```

```python
model.summary()
```

```python
model.layers
```

```python
hidden1 = model.layers[1]
hidde2 = model.layers[2]
```

```python
weights, biases = hidden1.get_weights()
```

```python
model.compile(loss="sparse_categorical_crossentropy",
             optimizer="adam",
             metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=20)
```

```python
pd.DataFrame(history.history).plot(figsize=(8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

```python
model.evaluate(X_test, y_test)
```

```python
predict_y = model.predict(X_test)
```

```python
predict_y
```

```python
predict_y.round(2)
```

```python
predict_class = np.argmax(predict_y, axis=1)
```

```python
predict_class
```

```python
item = randrange(10000)
plt.imshow(X_test[item])
print(f"Este é o numero {numbers[predict_class[item]]}")
```

```python
from sklearn import metrics
confusionmatrix = metrics.confusion_matrix(y_test, predict_class)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(confusionmatrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("The confusion matrix")
plt.ylabel("Actual number")
plt.xlabel("Classified number")
plt.xticks(np.arange(10) + 0.5, ["zero",
          "um",
          "dois",
          "tres",
          "quatro",
          "cinco",
          "seis",
          "sete",
          "oito",
          "nove",])
plt.yticks(np.arange(10) + 0.5, ["zero",
          "um",
          "dois",
          "tres",
          "quatro",
          "cinco",
          "seis",
          "sete",
          "oito",
          "nove",])
plt.show()
```

```python
image_number = 1

while os.path.isfile(f"new_digits/digit{image_number}.png"):
    try:
        img = cv2.imread(f"new_digits/digit{image_number}.png")[:,:,0]
        img = np.invert(np.array([img]))
        prediction = model.predict(img)
        print(f"This digit is probably {np.argmax(prediction)}")
        plt.imshow(img[0], cmap='gray')
        plt.show()
    except:
        print("Error")
    finally:
        image_number+=1
```
